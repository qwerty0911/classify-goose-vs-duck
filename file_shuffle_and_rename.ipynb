```python
import os

target='duck'
# 주어진 디렉토리에 있는 항목들의 이름을 담고 있는 리스트를 반환합니다.
# 리스트는 임의의 순서대로 나열됩니다.
file_path = './'+target
file_names = os.listdir(file_path)
file_names

```

```python
import random

for name in file_names:
    src = os.path.join(file_path, name)
    dst = target+'.'+str(random.randint(0, 1000000)) + '.jpg'
    dst = os.path.join(file_path, dst)
    os.rename(src, dst)
```

```python
file_path = './'+target
file_names = os.listdir(file_path)
file_names
```

```python
i = 0
for name in file_names:
    src = os.path.join(file_path, name)
    dst = target+'.'+str(i) + '.jpg'
    dst = os.path.join(file_path, dst)
    os.rename(src, dst)
    i += 1
```
